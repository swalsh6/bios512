```r
library('tidyverse')
```

```r
wind = read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-10-27/wind-turbine.csv',col_types=cols())
```

```r
wind %>% head
```

Each row in the table represents a single turbine.

Make a bar chart ðŸ“Š **sorted by frequency** of turbine count per province territory. (Hint: `fct_infreq` will sort your factor levels by frequency for subsequent plotting)

```r
wind_sorted = wind %>%
    mutate(prov_sorted = province_territory %>% fct_infreq() %>% fct_rev())

p = ggplot(wind_sorted, aes(y=prov_sorted)) +
    geom_bar() +
    theme_minimal() +
    labs(x = "Count", y = "Province/Territory") 

p
```

Make a bar chart **sorted by frequency** of turbine count per manufacturer. **Group the low frequency manufacturers into an `Other` category**.

**Bonus:** Can you force the `Other` category to be the last bar? 

```r
wind_sort2 = wind %>%
    mutate(manufacturer_sort = manufacturer %>% fct_lump(n = 6) %>% fct_infreq() %>% fct_relevel('Other', after = Inf) %>% fct_rev())
    
    
p = ggplot(wind_sort2, aes(y = manufacturer_sort)) +
    geom_bar() +
    theme_minimal() +
    labs(x = "Count", y = "Manufacturer") 

p
```
